# Running your own networks

1. Create the project structure and store your edge lists in data/edge_lists
2. Modify the "default_config.py" file (reccomend to name it "my_config.py" if you modify it)
3. Run the pipeline



## 1. Create the project structure

```python
from network_spatial_coherence import structure_and_args
config_path = "my_config.py"  # make sure this is the name of your configuration file
# This will create folders on the direcoty "data_dir". 
# Specify the path if you want a custom one, otherwise it wil be created in the code folder
dire = "/home/david/PycharmProjects/test_exp/"
args = structure_and_args.GraphArgs(override_config_path=config_path, data_dir=dire)  
```

## 2. Modify the configuration file
Modify the "default_config.py" file with your parameters. Change the name to "my_config.py" for clarity.

* Set *proximity_mode = 'experimental'*
* Set *title_edge_list = 'your_edge_list.csv'* or if it is a nx graph *title_edge_list = 'your_edge_list.pickle'*


```python
# structure_and_args.export_default_config('default_config.py')
```

## 3. Run the pipeline

If you just care about the reconstruction set the configuration file according to this:
*   'reconstruct': True,
*   'reconstruction_mode': 'STRND',
*    'spatial_coherence_validation': {   'gram_matrix': False,
                                        'network_dimension': False,
                                        'spatial_constant': False},
Make sure that the name of the configuration file matches the name you use in the code below.

```python
from network_spatial_coherence import nsc_pipeline

args.verbose = False
graph, args = nsc_pipeline.load_and_initialize_graph(args=args)
single_graph_args = nsc_pipeline.run_pipeline(graph, args)
nsc_pipeline.plot_profiling_results(single_graph_args)  

```
